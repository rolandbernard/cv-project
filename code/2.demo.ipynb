# Demo

This notebook contains the code used to generate demo sequences to show during the presentation of the project.

```python
import util
import dataset
import detect
import tracker
import kalman
```

```python
data = dataset.SalsaDataset()

physics = tracker.build_walled_physics(
    scale=1.0, center=(10, 3, 0), up=(0, 0, 1))
physics.to(util.DEVICE)

detector = detect.PoseDetector()
detector.to(util.DEVICE)

source = data.get_source()
source.to(util.DEVICE)

track = tracker.Tracker(detector, physics, mn_threshold=-5, mo_threshold=-2)
cams, frames, fps = track.evaluate(source, limit=4500)
util.save_tracks(f"./stats/demo.json", cams,
                 frames, fps, (10, 3, 0), (0, 0, 1))
```

```python
data = dataset.CmuPanopticDataset(path="./data/panoptic")

physics = tracker.build_constrained_physics()
physics.to(util.DEVICE)

detector = detect.PoseDetector()
detector.to(util.DEVICE)

source = data.get_source("160906_pizza1", num_vga_cams=8)
source.to(util.DEVICE)

track = tracker.Tracker(detector, physics)
cams, frames, fps = track.evaluate(source)
util.save_tracks(f"./stats/demo_pizza.json", cams, frames, fps)
```
